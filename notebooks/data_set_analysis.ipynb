# Post-processing

## Dates

```python
weekly_df['snapshot_dt'] = weekly_df['date'].map(lambda value: parser.parse(value).date())
```

```python
weekly_df['actual_start_dt'] = weekly_df['actual_start_date'].map(lambda value: parser.parse(value).date())
```

```python
weekly_df['created_dt'] = weekly_df['created'].map(lambda value: parser.parse(value).date())
```

```python
weekly_df['days_to_close'] = weekly_df.apply(lambda row: (row.actual_start_dt - row.snapshot_dt).days, axis=1)
```

```python
weekly_df['position_age'] = weekly_df.apply(lambda row: (row.snapshot_dt - row.created_dt).days, axis=1).astype(float)
```

```python
weekly_df['position_lifetime'] = weekly_df.apply(lambda row: (row.actual_start_dt - row.created_dt).days, axis=1).astype(float)
```

```python
weekly_df['period'] = weekly_df['actual_start_dt'].map(lambda value: int('%04d%02d' % (value.year, value.month)))
```

```python
weekly_df.period.unique()
```

```python
df = pd.DataFrame(listdir('../data/daily data set snaps/'), columns=['files'])

df['date'] = df['files'].map(lambda x: x.split('.')[0][-10:])

df['dates'] = pd.to_datetime(df['date'])
max_date = max(df['dates'])
```

```python
%%time
status_change = []

for i,f in enumerate(df['files']):
    try:
        f1 = f
        f2 = df['files'][i+1]
        
        df1 = pd.read_csv('../data/daily data set snaps/' + f1)
        df2 = pd.read_csv('../data/daily data set snaps/' + f2)
        
        df1 = df1[['position_id', 'staffing_status']]
        df2 = df2[['position_id', 'staffing_status']]

        df1.set_index('position_id', inplace=True)
        df2.set_index('position_id', inplace=True)
        
        df1 = df1.dropna()
        df2 = df2.dropna()
        
        temp = pd.merge(df1, df2, how = 'inner', left_index=True, right_index=True)
        temp['summ'] = (temp['staffing_status_x']==temp['staffing_status_y'])
        temp['first_next'] = temp['staffing_status_x'] + '_' + temp['staffing_status_y']
        
        status_change.append(pd.DataFrame(temp[temp['summ']==False].groupby('first_next')['summ'].count()).reset_index())
        
        print f1,f2
    except(KeyError):
        pass
```

```python
status_results = pd.concat(status_change)
```

```python
status_results = pd.DataFrame(status_results.groupby('first_next')['summ'].sum()).reset_index()
```

```python
status_results['first'] = status_results['first_next'].map(lambda x: x.split('_')[0])
status_results['next'] = status_results['first_next'].map(lambda x: x.split('_')[1])
```

```python
def count_multicat(series):
    massive = []
    for lists in series:
        for element in lists:
            massive.append(element)
    df_count = pd.DataFrame(massive, columns=['channels'])
    df_count['ind'] = df_count.index

    return df_count
```
